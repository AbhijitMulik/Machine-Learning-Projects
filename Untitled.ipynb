```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
housing= fetch_california_housing()
```

```python
housing
```

```python
df=pd.DataFrame(housing.data, columns=housing.feature_names)
```

```python
df
```

```python
df["Price"]=housing.target
```

```python
df
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_tree=DecisionTreeRegressor()
```

```python
model_tree.fit(X_train,y_train)
```

```python
model_tree.score(X_test,y_test) # score function gives R2 score for regression and accuracy score for classification
```

```python
model_tree.score(X_train,y_train)
```

```python
y_pred=model_tree.predict(X_test)
```

```python
#from sklearn.metrics import accuracy_score,classification_report - this is only for classification 
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("R² Score:", r2)
```

```python
import seaborn as sns
sns.displot(y_pred-y_test, kind='kde')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters= {
   'criterion': ['squared_error'],
'max_depth': [5, 10],
'min_samples_split': [2, 5],
'ccp_alpha': [0.0, 0.01]
}
```

```python
cv=GridSearchCV(model_tree,param_grid=parameters,cv=5,scoring="neg_mean_squared_error")
```

```python
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred=cv.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse=mean_squared_error(y_pred,y_test)
r2=r2_score(y_pred,y_test)
print(mse)
print(r2)
print(cv.best_score_)
```
