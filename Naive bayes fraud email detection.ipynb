```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("spam.csv")f
df.head()
```

```python
df.groupby("Category").describe()
```

```python
# df["spam"]=df["Category"].apply(lambda x:1 if x=="spam" else 0)
```

```python
df["Category"]=df["Category"].map({"ham":0,"spam":1})
```

```python
df
```

```python
X=df.Message
y=df.Category
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train= v.fit_transform(X_train)
X_test= v.transform(X_test)
X_train.toarray()[:3]
X_test.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
model = MultinomialNB()
parameter = {'alpha': [0.1, 0.5, 1.0, 2.0],'fit_prior': [True, False]
}
```

```python
nb_model = GridSearchCV(model, param_grid=parameter,scoring="accuracy",cv=5)
nb_model.fit(X_train, y_train)
```

```python
print(nb_model.best_params_)
```

```python
print(nb_model.best_score_)
```

```python
y_pred=nb_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
```

```python
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
emails = ["Hey Mohan, Can we go together to watch football match tomorrow ? ",
          "Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward" 
         ]
```

```python
email_count=v.transform(emails)
nb_model.predict(email_count)
```

```python
nb_model.score(X_test,y_test)
```
