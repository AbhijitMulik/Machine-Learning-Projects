```python

# Import libraries

from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# For Jupyter Notebook to display plots inline
%matplotlib inline

# Generate dataset (two interleaving half circles)
X, y = make_moons(n_samples=250, noise=0.05)

# View first few rows of X
print(X)

# Scatter plot of the raw dataset



```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.title("Original Data")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan.labels_, cmap='plasma')
plt.title("DBSCAN Clustering Results")
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='plasma')
plt.title("DBSCAN Clustering Results")
plt.show()
```
